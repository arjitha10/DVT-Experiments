```python
# Creating SQLite database

import pandas as pd
import sqlite3

# Read the CSV file
df = pd.read_csv(
    '/kaggle/input/datasets/imakash3011/customer-personality-analysis/marketing_campaign.csv',
    sep='\t'
)

# Create a SQLite database
conn = sqlite3.connect("marketing_campaign.db")

# Store the DataFrame as a table
df.to_sql("marketing_campaign", conn, if_exists="replace", index=False)

# Close the connection
conn.close()

print("SQLite database created successfully!")
```

```python
# Importing from database

import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("marketing_campaign.db")
print("Connection successful")

# Read data from the database table
df = pd.read_sql_query("SELECT * FROM marketing_campaign;", conn)

# Display the first 5 rows
print(df.head())

# Close the connection
conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv(
    "/kaggle/input/datasets/imakash3011/customer-personality-analysis/marketing_campaign.csv",
    sep="\t"
)

# Count customers by education level
education_count = df['Education'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8,5))
plt.bar(education_count.index, education_count.values)

plt.title("Distribution of Customers by Education")
plt.xlabel("Education")
plt.ylabel("Count")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "/kaggle/input/datasets/imakash3011/customer-personality-analysis/marketing_campaign.csv",
    sep="\t"
)

marital_count = df['Marital_Status'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(marital_count.index, marital_count.values)

plt.title("Distribution of Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)

plt.show()
```
